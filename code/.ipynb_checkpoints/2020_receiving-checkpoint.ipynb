```python
import pandas as pd
```

```python
df2020 = pd.read_csv('../data/2020_receiving_original.csv')
df2020['Year'] = 2020
df2020.head()
```

```python
df2019 = pd.read_csv('../data/2019_receiving_original.csv')
df2019['Year'] = 2019
df2019.head()
```

```python
simple2019 = df2019[['G', 'GS', 'Tgt', 'Rec', 'Yds', 'TD', 'Player-additional']].copy()
simple2019.rename(columns = {'G': 'G19', 'GS': 'GS19', 'Tgt': 'Tgt19', 'Rec': 'Rec19', 'Yds': 'Yds19', 'TD': 'TD19'}, inplace = True)
simple2019.head()
```

```python
temp = df2020.merge(simple2019, how = 'left', on = 'Player-additional')
temp.head()
```

```python
df2018 = pd.read_csv('../data/2018_receiving_original.csv')
df2018['Year'] = 2018
df2018.head()
```

```python
simple2018 = df2018[['G', 'GS', 'Tgt', 'Rec', 'Yds', 'TD', 'Player-additional']].copy()
simple2018.rename(columns = {'G': 'G18', 'GS': 'GS18', 'Tgt': 'Tgt18', 'Rec': 'Rec18', 'Yds': 'Yds18', 'TD': 'TD18'}, inplace = True)
simple2018.head()
```

```python
final2020 = temp.merge(simple2018, how = 'left', on = 'Player-additional')
final2020.head()
```

```python
final2020['Pos'] = final2020['Pos'].map({'wr': 'WR', 'te': 'TE', 'rb': 'RB', 'qb': 'QB', 'fb': 'FB', 
                                         'WR': 'WR', 'TE': 'TE', 'RB': 'RB', 'QB': 'QB', 'FB': 'FB', 'LT': 'LT', 'RT': 'RT'})
```

```python
final2020['Pos'].value_counts()
```

```python
#the Targets, Receptions, Yards, and TDs will be our target using the 2020 through 2018 stats
df2021 = pd.read_csv('../data/2021_receiving_original.csv')
df2021['Year'] = 2021
df2021.head()
```

```python
simple2021 = df2021[['Tgt', 'Rec', 'Yds', 'TD', 'Player-additional']].copy()
#removed Games and Games Started until I figure out what to do with those
simple2021.rename(columns = {'Tgt': 'Tgt21', 'Rec': 'Rec21', 'Yds': 'Yds21', 'TD': 'TD21'}, inplace = True) 
simple2021.head()
```

```python
target = final2020.merge(simple2021, how = 'inner', on = 'Player-additional')
print(target.shape)
target.head()
```
